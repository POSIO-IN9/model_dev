```python
import yaml

data = {
    "train": './train',
    "val": './val',
    "test": './test',
    "nc": 4 ,
    "names": ['fruit', 'buds', 'stem', 'leaf']
}

# YAML 파일에 데이터 쓰기
with open('./farmpj.yaml', 'w') as f:
    yaml.dump(data, f)

print("YAML 파일 생성 완료")


# 생성한 YAML 파일 내용 확인
with open('./farmpj.yaml', 'r') as f:
    lines = yaml.safe_load(f)
    print(lines)
```
